```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sine wave

```python
def generate_sine_wave_data():
    """Generate sine wave data."""
    x = np.linspace(0, 2*(2 * np.pi), 500,dtype=np.float32)
    y = np.sin(x)
    return x, y

x,y = generate_sine_wave_data()
sine = np.array([x,y]).T
np.savetxt('../data/sine.csv', sine, delimiter=',', )
plt.scatter(sine[:,0], sine[:,1])
```
